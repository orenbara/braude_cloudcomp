```python
# @title Fetch - get website data
import requests
from bs4 import BeautifulSoup

def fetch_page(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        return soup
    else:
        return None
```

```python
# @title create Index - python dict with alphanumerical words
import re

def index_words(soup):
    index = {}
    words = re.findall(r'\w+', soup.get_text())
    for word in words:
        word = word.lower()
        if word in index:
            index[word] += 1
        else:
            index[word] = 1
    return index
```

```python
# @title Remove stop words
def remove_stop_words(index):
    stop_words = {'a', 'an', 'the', 'and', 'or', 'in', 'on', 'at', 'to'}
    for stop_word in stop_words:
        if stop_word in index:
            del index[stop_word]
    return index
```

```python
# @title Apply stemming with Porter algorithm
from nltk.stem import PorterStemmer

def apply_stemming(index):
    stemmer = PorterStemmer()
    stemmed_index = {}
    for word, count in index.items():
        stemmed_word = stemmer.stem(word)
        if stemmed_word in stemmed_index:
            stemmed_index[stemmed_word] += count
        else:
            stemmed_index[stemmed_word] = count
    return stemmed_index
```

```python
# @title Search - provide query and index for a page
def search(query, index):

    stemmer = PorterStemmer()
    query_words = re.findall(r'\w+', query.lower())
    print(query)
    results = {}
    for word in query_words:
        word = stemmer.stem(word)
        if word in index:
            results[word] = index[word]
    return results
```

```python
# @title Client function for searching
def search_engine(url, query):
    soup = fetch_page(url)
    if soup is None:
        return None
    index = index_words(soup)
    index = remove_stop_words(index)
    index = apply_stemming(index)
    results = search(query, index)
    return results
```

```python
# @title User input for search engine
url = 'https://en.wikipedia.org/wiki/Bird' # @param {type:"string"}
query = 'bird' # @param {type:"string"}
results = search_engine(url, query)
print(results)
```

```python
# @title Rank previous result
rank=1
for word, count in results.items():
  rank = rank*1/count
rank = 1-rank
print(rank)
```

```python
# @title EX1 - Braude English
url = 'https://w3.braude.ac.il/?lang=en' # @param {type:"string"}
query = 'industry Braude college Galilee center' # @param {type:"string"}
results = search_engine(url, query)
print(results)
rank=1
for word, count in results.items():
  rank = rank*1/count
rank = 1-rank
print(rank)
```

EX 1 Part 2 - new ranking system: we will use addition instead of multiplication. when using addition the rank will not be effected as much by high repetition of words - which not always is indication of relevancy.


```python
# @title EX1 part 2 implementation
url = 'https://w3.braude.ac.il/?lang=en' # @param {type:"string"}
query = 'industry Braude college Galilee center' # @param {type:"string"}
results = search_engine(url, query)
print(results)
rank=1
for word, count in results.items():
  rank += count

print(rank)
```

Ex1 part 3: generaly, the first method will provide more relevant result sicne repetition of a word suggests that is is a dominant topic in the page and multiplication will make sure that when a word is repeated it will score many points in the ranking system.
